```r
library(tidyverse)
library(janitor)
library(psych)
survey <- read_csv("/kaggle/input/diabetes-dataset/diabetes.csv")
```

```r
describe(survey)
describe.by(survey, survey$Glucose)
```

```r
hist(survey$Glucose)
hist(survey$BMI)
```

```r
barplot(table(survey$Glucose))
barplot(table(survey$BMI))
barplot(table(survey$BMI, survey$Glucose))
barplot(table(survey$BMI,survey$Glucose),
beside=TRUE)

```

```r
plot(survey$Glucose,survey$BMI)
cor(survey$BMI,survey$Glucose)
Model <- lm(BMI ~ Glucose, data = survey)
```

```r
plot(survey$BMI ~ survey$Glucose, xlab="glucose", ylab="bmi")
abline(Model, col="blue")
```

```r
newdata <- data.frame(Glucose=100)
predict(Model,newdata)
```
